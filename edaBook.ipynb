GLOBAL IMPORTS AND CONFIGS

```python
import  os
import  warnings

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import  plotly.express as px
import seaborn as sns
import data_profiling

#IGNORE WARNINGS
warnings.filterwarnings("ignore")
```

LOADING THE DATA AND REVIEW

```python
DATA_PATH = os.path.join("dataset", "Dataset of Diabetes .csv")

df = pd.read_csv(DATA_PATH)

print("Dataset Shape:", df.shape)

df.head()
```

CHECKING FOR MISSING VALUES

```python
df.info()
```

CHECKING FOR DUPES

```python
duplicates = df.duplicated().sum()

print("Total Duplicate Rows:", duplicates)
```

PROFILING THE DATASET WITH YDATAPROFILER

```python
profile = data_profiling.ProfileReport(
    df,
    title="Diabetes Dataset Profiling Report",
    explorative=True
)

profile.to_notebook_iframe()
```

CHECKING THE CLASS DISTRIBUTION (TARGET VARIABLE DISTRIBUTION)

```python
plt.figure(figsize=(7, 5))

sns.countplot(
    x="CLASS",
    data=df,
    order=["N", "P", "Y"]
)

plt.title("Distribution of Diabetes Classes")

plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```
